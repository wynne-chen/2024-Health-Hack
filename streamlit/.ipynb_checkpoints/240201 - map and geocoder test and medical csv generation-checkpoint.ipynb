```python
# Create a geocoder instance
geolocator = Nominatim(user_agent="my_geocoder")

# Input field for address
address = input("Enter an address in Singapore: ")
latitude = 0.0
longitude = 0.0

if True:
    if address:
        address = address.strip().lower()
        if 'blk' in address:
            address = address.replace('blk', '').strip()
        elif re.search('([s])?(\d{6})', address):
            address = address.replace('s', '')
            
        try:
            location = geolocator.geocode(query = address, timeout = 5, addressdetails = True)
            
            country = location.raw['address']['country']
            
            if country != 'Singapore':
                raise ValueError('Please enter an address in Singapore.')

            if location:
                latitude = location.latitude
                longitude = location.longitude

                print(f"The neighbourhood is: {location.raw['suburb']}. Latitude: {latitude}, Longitude: {longitude}")
            else:
                print("Address not found.")
        except Exception as e:
            print("Address not found.")
    else:
        print("Please enter a valid address.")
```
